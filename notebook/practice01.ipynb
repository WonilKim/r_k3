```r
gsub("X", "", paste(n[1], n[2], "전체", sep="."))
# "X"를 찾아서 ""로 바꾸고 n[1], n[2], "전체" 을 "."을 붙여서 이어라.
```

```r
# f <- function(x) {
#     n <- strsplit(x, "\\.")
#     return(n)
# }

f <- function(x) {
    n <- unlist(strsplit(x, "\\."))
    if (length(n) == 1) {
        return(x)
    } else if (length(n) == 2) {
        return(gsub("X", "", paste(n[1], n[2], "전체", sep=".")))
    } else {
        if (identical(n[3], "1")) { # n[3] == 1
            return(gsub("X", "", paste(n[1], n[2], "남자", sep=".")))
        } else {
            return(gsub("X", "", paste(n[1], n[2], "여자", sep=".")))
        }
    }
    return(n)
}
```

```r
names(df) <- lapply(colnames(df), f)
```

### 4. 데이터 정리

```r
# variable 컬럼의 데이터를 원자단위로 나눠야 필요한 쿼리를 날릴 수 있다.
# 예: 1997년의 전국 데이터를 가져오려는 경우.
```

```r
f1 <- function(x) {
    n <- unlist(strsplit(x, "\\."))
    return(n[1])
}
f2 <- function(x) {
    n <- unlist(strsplit(x, "\\."))
    return(n[2])
}
f3 <- function(x) {
    n <- unlist(strsplit(x, "\\."))
    return(n[3])
}
```

```r
df2["연도"] <- apply(df2["variable"], 1, f1)
df2["월"] <- apply(df2["variable"], 1, f2)
df2["성별"] <- apply(df2["variable"], 1, f3)
head(df2)
```

```r
colnames(df2)

colnames(df2)[3] <- "출생아수"
head(df2)
```
